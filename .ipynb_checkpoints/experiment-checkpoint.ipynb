```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, log_loss
```

```python
def count_frequency(df):
    freq_dict = {'Sunday':0,'Monday':0,'Tuesday':0,'Wednesday':0,'Thursday':0,'Friday':0,'Saturday':0}
    for i in df[column_name]:
        freq_dict[i] += 1
    return freq_dict
```

```python
def convert_categorical(df):
    day_dict = {'Sunday':0,'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6}
    district_dict = {'NORTHERN':0, 'PARK':1, 'INGLESIDE':2, 'BAYVIEW':3, 'RICHMOND':4, 
                     'CENTRAL':5, 'TARAVAL':6,'TENDERLOIN':7, 'MISSION':8, 'SOUTHERN':9}
    df['DayOfWeek'] = df['DayOfWeek'].replace(day_dict)
    df['PdDistrict'] = df['PdDistrict'].replace(district_dict)
    return df
```

```python
df = pd.read_csv('data/train.csv')
```

```python
target = df['Category'].unique()
print(target)
```

```python
freq = count_frequency(df, 'DayOfWeek')
plt.bar(range(len(freq)), freq.values(), align='center')
plt.xticks(range(len(freq)), freq.keys())
plt.show()
```

```python
X = df.drop(['Category','Descript','Resolution'], axis=1)
X['Dates'] = pd.to_datetime(X['Dates'])
X.loc[:,'X':'Y'] = X.loc[:,'X':'Y'].apply(pd.to_numeric)
X = convert_categorical(X)
y = df['Category']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
sns.lmplot(x='X',y='Y',data=df)
```

```python
features = ['DayOfWeek','PdDistrict','X','Y']
```

```python
clf = AdaBoostClassifier()
clf.fit(X_train[features], y_train)
```

```python
pred = clf.predict(X_test[features])
```

```python
accuracy_score(pred, y_test)
```
